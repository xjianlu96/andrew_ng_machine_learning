forward propagation

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def forward_propagate(X,theta1,theta2):
    m=X.shape[0]
    a1=np.insert(X,0,values=np.ones(m),axis=1)
    z2=a1*theta1.T
    a2=np.insert(sigmoid(z2),0,values=np.ones(m),axis=1)
    z3=a2*theta2.T
    h=sigmoid(z3)
    return a1,z2,a2,z3,h
```

```python
def cost(theta1,theta2,input_size,hidden_size,num_labels,X,y,learning_rate):
    m=X.shape[0]
    X=np.matrix(X)
    y=np.matrix(y)
    
    a1,z2,a2,z3,h=forward_propagate(X,theta1,theta2)
    
    J=0
    for i in range(m):
        first=np.multiply(-y[i,:],np.log(h[i,:]))
        second=np.multiply((1-y[i,:]),np.log(1-h[i,:]))
        J=J+np.sum(first-second)
    J=J/m
    
    return J
```

backward propagation

```python
def sigmoid_gradient(z):
    return np.multiply(sigmoid(z),(1-sigmoid(z)))
```

```python
sigmoid_gradient(0)
```

```python
#set initial epslion to 0.12 so the interval of epslion is [-0.12,0.12]
#np.random.random(3)
params = (np.random.random(size=hidden_size * (input_size + 1) + num_labels * (hidden_size + 1)) - 0.5) * 0.24
```

```python
def backprop(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    m=X.shape[0]
    X=np.matrix(X)
    y=np.matrix(y)
    
    #run the foward propagation
    a1,z2,a2,z3,h=forward_propagate(X,theta1,theta2)
    
    #reshape the parameter array into parameter
    theta1=np.matrix(np.reshape(params[:hidden_size * (input_size + 1)],(hidden_size,(input_size+1)))) #(25,401)
    theta2=np.matrix(np.reshape(params[hidden_size * (input_size + 1):],(num_labels * (hidden_size+1)))) #(10*26)
    
    #initializations
    J=0
    delta1=np.zeros(theta1.shape) #(25,401)
    delta2=np.zeros(theta2.shape) #(10,26)
    
    #compute the cost
    for i in range(m):
        first=np.multiply(-y[i,:],np.log(h[i,:]))
        second=np.multiply((1-y[i,:]),np.log(1-h[i,:]))
        J=J+np.sum(first-second)
    J=J/m
    
    #the cost regularization term
    J=J+(float(learning_rate)/(2*m))*(np.sum(np.power(theta1[:,1:],2))+np.sum(np.power(theta2[:,1:],2)))
    
    #perform backward propagation
    for t in range(m):
        a1t=a1[t,:] #(1,401)
        z2t=z2[t,:] #(1,25)
        a2t=a2[t,:] #(1,26)
        ht=h[t,:] #(1,10)
        yt=y[t,:] #(1,10)
        
        d3t=ht-yt #(1,10)
        
        z2t=np.insert(z2t,0,values=np.ones(1)) #(1,26)
        d2t=np.multiply((theta2.T*d3t.T).T,sigmoid_gradient(z2t)) #(1,26)
        
        delta1=delta1+(d2t[:,1:]).T*a1t
        delta2=delta2+d3t.T*a2t
    
    delta1=delta1/m
    delta2=delta2/m
    
    return J,delta1,delta2
```

```python
def backpropReg(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    m=X.shape[0]
    X=np.matrix(X)
    y=np.matrix(y)

    #reshape the parameter array into parameter
    theta1=np.matrix(np.reshape(params[:hidden_size * (input_size + 1)],(hidden_size,(input_size+1))))
    theta2=np.matrix(np.reshape(params[hidden_size * (input_size + 1):],(num_labels,(hidden_size+1))))
    
    #run the foward propagation
    a1,z2,a2,z3,h=forward_propagate(X,theta1,theta2)
    
    #initializations
    J=0
    delta1=np.zeros(theta1.shape) #(25,401)
    delta2=np.zeros(theta2.shape) #(10,26)
    
    #compute the cost
    for i in range(m):
        first=np.multiply(-y[i,:],np.log(h[i,:]))
        second=np.multiply((1-y[i,:]),np.log(1-h[i,:]))
        J=J+np.sum(first-second)
    J=J/m
    
    #perform backward propagation
    for t in range(m):
        a1t=a1[t,:] #(1,401)
        z2t=z2[t,:] #(1,25)
        a2t=a2[t,:] #(1,26)
        ht=h[t,:] #(1,10)
        yt=y[t,:] #(1,10)
        
        d3t=ht-yt #(1,10)
        
        z2t=np.insert(z2t,0,values=np.ones(1)) #(1,26)
        d2t=np.multiply((theta2.T*d3t.T).T,sigmoid_gradient(z2t)) #(1,26)
        
        delta1=delta1+(d2t[:,1:]).T*a1t
        delta2=delta2+d3t.T*a2t
        
    delta1=delta1/m
    delta2=delta2/m
    
    # add the gradient regularization term
    delta1[:,1:] = delta1[:,1:] + (theta1[:,1:] * learning_rate) / m
    delta2[:,1:] = delta2[:,1:] + (theta2[:,1:] * learning_rate) / m
    
    grad = np.concatenate((np.ravel(delta1), np.ravel(delta2)))
    
    return J,grad
```

```python
from scipy.optimize import minimize

fmin = minimize(fun=backpropReg, x0=(params), args=(input_size, hidden_size, num_labels, X, y_onehot, learning_rate), 
                method='TNC', jac=True, options={'maxiter': 250})

fmin
```
