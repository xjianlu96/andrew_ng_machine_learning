```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from scipy.io import loadmat
from sklearn.metrics import classification_report
```

```python
weight=loadmat("E:/andrew machine learning/andrew_ml_ex33507/ex3weights.mat")
theta1,theta2=weight["Theta1"],weight["Theta2"]
theta1.shape,theta2.shape
```

```python
data=loadmat("E:/andrew machine learning/andrew_ml_ex33507/ex3data1.mat")
data["X"].shape,data["y"].shape
```

```python
sample_idx=np.random.choice(np.arange(data['X'].shape[0]), 100)
sample_images=data['X'][sample_idx, :]
sample_images
```

```python
fig, ax_array = plt.subplots(nrows=10, ncols=10, sharey=True, sharex=True, figsize=(12, 12))
for r in range(10):
    for c in range(10):
        ax_array[r, c].matshow(np.array(sample_images[10 * r + c].reshape((20, 20))).T,cmap=matplotlib.cm.binary)
        plt.xticks(np.array([]))
        plt.yticks(np.array([]))
```

```python
X = np.insert(data['X'], 0, values=np.ones(data["X"].shape[0]), axis=1)
```

```python
X2=np.matrix(np.insert(data["X"],0,values=np.ones(X.shape[0]),axis=1))
```

```python
y2=np.matrix(data["y"])
```

```python
X2.shape,y2.shape
```

```python
a1=X2
z2=a1*theta1.T
z2.shape
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
a2=sigmoid(z2)
a2.shape
```

```python
a2=np.insert(a2,0,values=np.ones(a2.shape[0]),axis=1)
z3=a2*theta2.T
z3.shape
```

```python
a3=sigmoid(z3)
```

```python
y_pred=np.argmax(a3,axis=1)+1
y_pred.shape
```

```python
print(classification_report(y2,y_pred))
```
